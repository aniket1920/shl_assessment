```python
import numpy as np
import pandas as pd
from app.embeddings.embedder import CatalogEmbedder
from app.retrieval.faiss_retriever import FaissRetriever
```

```python
df = pd.read_csv("../data/processed/catalog_processed.csv")
```

```python
embeddings = np.load(
    "../data/embeddings/catalog_embeddings.npy"
)
```

```python
retriever = FaissRetriever(
    embeddings
)
```

```python
embedder = CatalogEmbedder()
```

```python
query = "Java Spring SQL"
query_embedding = embedder.encode([query])
```

```python
results = retriever.search(
    query_embedding,
    top_k=5
)
```

```python
for result in results:
    row = df.iloc[result["index"]]
    print("=" * 80)
    print(row["name"])
    print(row["assessment_type"])
    print(result["score"])
```
