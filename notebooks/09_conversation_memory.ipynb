```python
conversation = [
    {
        "role":"user",
        "content":"Need assessment"
    },
    {
        "role":"assistant",
        "content":"Could you tell me the role?"
    },
    {
        "role":"user",
        "content":"Graduate Java Engineer"
    }
]
```

```python
def latest_user_message(conversation):
    for message in reversed(conversation):
        if message["role"] == "user":
            return message["content"]
    return ""
```

```python
latest_user_message(conversation)
```

```python
def previous_user_messages(conversation):
    messages = []
    for message in conversation:
        if message["role"] == "user":
            messages.append(message["content"])
    return messages
```

```python
previous_user_messages(conversation)
```

```python
def build_context(conversation):
    messages = previous_user_messages(conversation)
    return "\n".join(messages)
```

```python
print(build_context(conversation))
```

```python
def build_context(conversation):
    lines = []
    for message in conversation:
        role = message["role"].capitalize()
        content = message["content"]
        lines.append(f"{role}: {content}")
    return "\n".join(lines)
```

```python
print(build_context(conversation2))
```
