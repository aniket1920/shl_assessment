```python
import sys
from pathlib import Path
PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))
```

```python
from app.catalog.loader import load_catalog
import pandas as pd
```

```python
catalog = load_catalog()
df = pd.DataFrame(catalog)
```

```python
from app.catalog.preprocess import (
    build_search_document,
    extract_duration_minutes,
    is_report,
)
```

```python
df["search_document"] = [
    build_search_document(a)
    for a in catalog
]
df["duration_minutes"] = df["duration"].apply(extract_duration_minutes)
df["is_report"] = df["name"].apply(is_report)
```

```python
df[
    [
        "name",
        "duration",
        "duration_minutes",
        "is_report"
    ]
].head(15)
```

```python
df["assessment_type"] = df["keys"].apply(
    lambda x: ", ".join(x)
)

df["job_level"] = df["job_levels"].apply(
    lambda x: ", ".join(x)
)

df["language"] = df["languages"].apply(
    lambda x: ", ".join(x)
)
```

```python
df[
    [
        "name",
        "assessment_type",
        "job_level",
        "language"
    ]
].head()
```

```python
from pathlib import Path
output = Path("../data/processed/catalog_processed.csv")
df.to_csv(output, index=False)
```
