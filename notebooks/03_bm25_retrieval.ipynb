```python
import pandas as pd
df = pd.read_csv("../data/processed/catalog_processed.csv")
df.head()
```

```python
from app.retrieval.bm25 import BM25Retriever
```

```python
documents = df["search_document"].tolist()
retriever = BM25Retriever(documents)
```

```python
results = retriever.search(
    "Java Spring SQL",
    top_k=10
)
results
```

```python
for result in results:
    row = df.iloc[result["index"]]
    print("=" * 80)
    print(row["name"])
    print(result["score"])
```

```python
results = retriever.search(
    "Java Spring SQL",
    top_k=5
)
for result in results:
    row = df.iloc[result["index"]]
    print("=" * 80)
    print(f"Name  : {row['name']}")
    print(f"Type  : {row['assessment_type']}")
    print(f"Score : {result['score']:.3f}")
```
